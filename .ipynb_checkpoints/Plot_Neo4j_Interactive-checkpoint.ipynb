```python
from neo4j import GraphDatabase
import networkx as nx
```

# Set up Connection to Neo4j

```python
URI = "bolt://localhost:7687"
USER = "neo4j"
PASSWORD = "12345678"
```

```python
driver = GraphDatabase.driver(URI, auth=(USER, PASSWORD))
```

# Build a NetworkX graph from Neo4j DB

```python
def load_graph(driver, limit=500):
    query = """
    MATCH (p:Person)-[:POSTED]->(t:Tweet)-[:MENTIONS]->(s:Stock)
    RETURN p, t, s
    LIMIT $limit
    """

    G = nx.DiGraph()

    with driver.session() as session:
        result = session.run(query)

        for record in result:
            p = record["p"]
            t = record["t"]
            s = record["s"]

            # Node IDs
            p_id = f"person_{p.id}"
            t_id = f"tweet_{t.id}"
            s_symbol = f"stock_{s.symbol}"

            # --- Add nodes ---
            G.add_node(
                p_id,
                label=p.get("name", "Unnamed Person"),
                type="Person",
                **dict(p)
            )
            G.add_node(
                t_id,
                label=f"Tweet {t.id}",
                type="Tweet",
                **dict(t)
            )
            G.add_node(
                s_symbol,
                label=s.get("symbol", "Unknown Stock"),
                type="Stock",
                **dict(s)
            )

            # --- Add edges ---
            # Person → Tweet
            G.add_edge(
                p_id,
                t_id,
                label="POSTED"
            )

            # Tweet → Stock
            G.add_edge(
                t_id,
                s_symbol,
                label="MENTIONS"
            )

    return G
```

```python
load_graph(driver)
```
